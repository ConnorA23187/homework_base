1. 以下程序的功能是对随机产生的 1～100 之间的一个整数 n，利用二分法查找这个数并输出查找的次数。

```python
import random
import math
n = random.randint(1, 100)  # 函数 random.randint(a,b) 返回从 a 到 b 之间随机选出的一个整数
a, b, cnt = 1, 100, 0
while a < b:
    cnt += 1
    m = math.ceil((a + b) / 2)  # 函数 math.ceil(x) 的返回值为数值 x 的向上取整
    if n < m:
        b = m - 1
    elif n > m:
        a = m + 1
    else:
        print(n, cnt)
        break
```

2. 以下程序读入正整数 n，计算
 S = 1! − 2! + 3! − 4! + 5! − …… ± n!
并输出.

```python
n = int(input())
S = 0
sign = 1
k = 1
for i in range(1, n + 1):
    k = k * i
    S += sign * k
    sign = sign * -1
print(S)
```

3. 下面程序中的 Change 函数的功能是将十进制数转化 N 进制数（2 <= N <= 16）。
 输入目标进制和待转换的数，函数返回转换后的目标进制数。

```python
def Change(radix, num):
    g = []
    while num != 0:
        r = num % radix
        num = num // radix
        g.append(r)
    ans = ''
    t = ('A', 'B', 'C', 'D', 'E', 'F')
    for x in range(len(g)):
        if g[x] >= 10:
            ans += t[g[x] - 10]
        else:
            ans += str(g[x])
    return ans[::-1]
print(Change(16, 12455))
```

4. 给定一个整数的数字列表 L，判断 L 中是否存在相同数字，若存在返回 True，否则返回 False。


```python
def identical(L):
    flag = 0
    L.sort()
    for i in range(len(L) - 1):
        if L[i] == L[i + 1]:
            flag = 1
            break
    if flag:
        return True
    else:
        return False
```

Python数据分析（Pandas）
1)
描述（Description）
你获得了一组包含出租车司机及其乘车信息的数据集。
 你的任务是对这些数据进行一些基本分析，并将结果保存到一个 CSV 文件中。

```python
import pandas as pd
drivers = pd.read_csv('data/driver.csv')

rides_group = list()
for i in range(1, 5):
    rides_group.append(pd.read_csv(f'data/rides_{i}.csv'))
rides = pd.concat(rides_group)

result = dict()

def get_Average_Driver_Rating(drivers):
    return round(drivers['rating'].mean(), 2)

result['average_driver_rating'] = get_Average_Driver_Rating(drivers)

def get_Percentage_of_Drivers_with_Second_Language(drivers):
    return round(len(drivers['second_language'].dropna()) / len(drivers), 2)

result['percentage_of_drivers_with_second_language'] = get_Percentage_of_Drivers_with_Second_Language(drivers)

def get_Ride_Success_Rate(rides):
    return round(len(rides[rides['status'] == '成功']) / len(rides), 2)

result['ride_success_rate'] = get_Ride_Success_Rate(rides)

result_form = pd.DataFrame({
    "insight_type" : result.keys(),
    "values" : result.values()
})
print(result_form)
result_form.to_csv('result/analysis_results.csv')
```

深度学习训练相关名词解释
1.简述 Batch Normalization 的基本原理与作用。
批量归一化是在训练过程中进行归一化的方法,根据每个批次输入的数据的均值和方差进行归一化,最后将数据整体综合考虑,得到训练结束后使用的参数,它可以帮助神经网络在训练过程中减少梯度消失或梯度爆炸的出现。

2.什么是 Dropout？它在深度神经网络训练中的作用是什么？
Dropout 是一种训练时采用的正则化方法,通过设置一定概率使某些神经元的输出变为0(失效),来使模型获得更好的效果

3.解释梯度裁剪的概念及其在训练过程中的作用。
梯度裁剪是训练过程中对单次梯度设置阈值,当单次梯度绝对值超过阈值时,将单次梯度裁剪为阈值大小,以此减少梯度消失和梯度爆炸的风险

4.什么是早停？为什么它能防止过拟合？
早停通过监测训练过程中的指标控制模型是否结束训练,通常会设置一个阈值当连续多次(耐心值),该指标的改善值小于阈值时,结束训练;通过限制训练次数防止过拟合

5.数据增强（Data Augmentation）在训练过程中的意义是什么？请举例说明。
数据增强是指通过人工扩展训练集，增强模型的泛化能力、稳定性, 通常在某类数据较少时,使用一系列方法扩充数据集,使模型更健壮

6.什么是梯度消失？ 为什么会出现？ 有哪些解决办法？
梯度消失是指梯度随训练次数增加的时候逐渐变小,导致无法有效的改善模型参数,难以获得更好的效果
问题在于一些激活函数存在数值上限且分布不均匀,如sigmoid函数在数值超过一定程度后,其结果都大体趋近于1
通常可以使用改善的激活函数,更科学的归一化,对数据进行归一化来改善梯度消失的问题

7.什么是学习率调度？有什么好处？
学习率调度是会在训练过程中让学习率根据指标动态调整学习率, 达到优化训练的目的,使模型能够更好的跳出局部坑点,更快的找到优解


构建5层Sequential神经网络预测股票价格，要求包含：
●2个LSTM层（首层units=4需设置input_shape）
●每个LSTM层后接Dropout层(rate=0.2)
●最终Dense层(units=1) 实现步骤：
1.使用stock_model变量构建模型
2.编译模型(loss=‘mean_squared_error’, optimizer=‘adam’)
3.训练配置(epochs=5, batch_size=16, verbose=0)
4.输出模型summary()
5.预测x_test并输出最后一列
注意：x_train, y_train, x_test不用再写代码改动

```python
import tensorflow as tf
import numpy as np
# 假设我们用 60 天的价格预测第 61 天价格
time_steps = 60
features = 1
samples = 200

# 构造输入数据 (samples, time_steps, features)
x_train = np.random.rand(samples, time_steps, features)  # 200个 60天股票价格
y_train = np.random.rand(samples, 1)                     # 200个 对应的第61天股票价格
x_test  = np.random.rand(10, time_steps, features)       # 测试

stock_model = tf.keras.models.Sequential([    
    tf.keras.layers.LSTM(units=4, input_shape=(time_steps, features), return_sequences=True),
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.LSTM(units=4),
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.Dense(units=1)]
)
stock_model.compile(loss='mean_squared_error', optimizer='adam')
stock_model.fit(x_train, y_train, epochs=5, batch_size=16, verbose=0)
stock_model.summary()
```
